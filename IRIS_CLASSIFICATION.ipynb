```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.metrics import accuracy_score

# Load the dataset
file_path = '/content/IRIS.csv'
iris_data = pd.read_csv(file_path)

# Display basic information about the dataset
print("Dataset Information:")
print(iris_data.info())

# Check for any missing values
print("\nMissing Values in Each Column:")
print(iris_data.isnull().sum())

# Preview the first few rows of the dataset
print("\nFirst 5 Rows of the Dataset:")
print(iris_data.head())

# Visualize the distribution of species
plt.figure(figsize=(6,6))
sns.countplot(x='species', data=iris_data)
plt.title('Count of Each Iris Species')
plt.show()

# Encode species column to numerical values
species_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
iris_data['species'] = iris_data['species'].map(species_mapping)

# Plot a heatmap of the correlation matrix
plt.figure(figsize=(5,5))
sns.heatmap(iris_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Split the dataset into features and target variable
X_features = iris_data.drop(columns='species')
Y_target = iris_data['species']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_features, Y_target, test_size=0.1, stratify=Y_target, random_state=42)

# Initialize classifiers
logistic_model = LogisticRegression(max_iter=200)
decision_tree = DecisionTreeClassifier()
knn_classifier = KNeighborsClassifier()
naive_bayes = GaussianNB()
svm_classifier = svm.SVC(kernel='linear')

# Train the classifiers
logistic_model.fit(X_train, Y_train)
decision_tree.fit(X_train, Y_train)
knn_classifier.fit(X_train, Y_train)
naive_bayes.fit(X_train, Y_train)
svm_classifier.fit(X_train, Y_train)

# Evaluate the classifiers
models = {
    "Logistic Regression": logistic_model,
    "Decision Tree": decision_tree,
    "K-Nearest Neighbors": knn_classifier,
    "Naive Bayes": naive_bayes,
    "Support Vector Machine": svm_classifier
}

for model_name, model in models.items():
    train_predictions = model.predict(X_train)
    test_predictions = model.predict(X_test)
    train_accuracy = accuracy_score(Y_train, train_predictions)
    test_accuracy = accuracy_score(Y_test, test_predictions)
    print(f"{model_name} - Training Accuracy: {train_accuracy:.2f}, Test Accuracy: {test_accuracy:.2f}")

# Example prediction using one of the classifiers
example_data = np.array([[5.7, 2.5, 5.0, 2.0]])
example_data_reshaped = example_data.reshape(1, -1)

print("\nPrediction for Example Data:")
for model_name, model in models.items():
    prediction = model.predict(example_data_reshaped)
    print(f"{model_name}: Predicted class - {prediction[0]}")

# Display the first few test samples
print("\nFirst 5 Test Samples:")
print(X_test.head())
```

```python
import gradio as gr
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
from sklearn.metrics import accuracy_score

# Load the dataset
iris_data = pd.read_csv('/content/IRIS.csv')

# Encode species column to numerical values
species_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
iris_data['species'] = iris_data['species'].map(species_mapping)

# Split the dataset into features and target variable
X_features = iris_data.drop(columns='species')
Y_target = iris_data['species']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X_features, Y_target, test_size=0.1, stratify=Y_target, random_state=42)

# Initialize classifiers
logistic_model = LogisticRegression(max_iter=200)
decision_tree = DecisionTreeClassifier()
knn_classifier = KNeighborsClassifier()
naive_bayes = GaussianNB()
svm_classifier = svm.SVC(kernel='linear')

# Train the classifiers
logistic_model.fit(X_train, Y_train)
decision_tree.fit(X_train, Y_train)
knn_classifier.fit(X_train, Y_train)
naive_bayes.fit(X_train, Y_train)
svm_classifier.fit(X_train, Y_train)

# Models dictionary
models = {
    "Logistic Regression": logistic_model,
    "Decision Tree": decision_tree,
    "K-Nearest Neighbors": knn_classifier,
    "Naive Bayes": naive_bayes,
    "Support Vector Machine": svm_classifier
}

# Gradio function to make predictions
def predict_species(sepal_length, sepal_width, petal_length, petal_width, model_name):
    example_data = np.array([[sepal_length, sepal_width, petal_length, petal_width]])
    model = models[model_name]
    prediction = model.predict(example_data)
    species_reverse_mapping = {0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'}
    return species_reverse_mapping[prediction[0]]

# Gradio interface
with gr.Blocks() as demo:
    gr.Markdown("## Iris Species Classifier")
    sepal_length = gr.Number(label="Sepal Length")
    sepal_width = gr.Number(label="Sepal Width")
    petal_length = gr.Number(label="Petal Length")
    petal_width = gr.Number(label="Petal Width")
    model_name = gr.Dropdown(choices=list(models.keys()), label="Model")
    result = gr.Textbox(label="Predicted Species")

    btn = gr.Button("Predict")
    btn.click(predict_species, inputs=[sepal_length, sepal_width, petal_length, petal_width, model_name], outputs=result)

demo.launch()

```
